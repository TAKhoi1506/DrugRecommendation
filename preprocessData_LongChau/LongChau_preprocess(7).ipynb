```python
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
```

```python
data_LC = pd.read_csv('D:\OU\DoAn\DrugRecommandation\crawlData_LongChau\LongChau_rawData.csv',encoding='utf-8')
print(f'Kích thước dữ liệu ban đầu: {data_LC.shape}')
print(f'Các cột trong dữ liệu: {data_LC.columns.tolist()}')
print('Thông tin tổng quan')
print(data_LC.info())
data_LC 
```

<h3>Xử lý giá trị thiếu</h3>

```python
missing_values = data_LC.isnull().sum()
print("Giá trị thiếu trong mỗi cột:")
print(missing_values)

def handle_missing_values(df): # Xử lý giá trị thiếu và các thông tin "Đang cập nhật"
    missing_indicators = ['Đang cập nhật', 'đang cập nhật', 'ĐANG CẬP NHẬT',
        'Liên hệ', 'liên hệ', 'LIÊN HỆ',
        'Chưa có thông tin', 'chưa có thông tin',
        'N/A', 'n/a', 'null', 'NULL',]

    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].replace(missing_indicators, np.nan)
            df[col] = df[col].astype(str).str.strip()
            df[col] = df[col].replace('nan', np.nan)
    return df
data_LC= handle_missing_values(data_LC)
print('Số lượng giá trị thiếu sau khi xử lý:')
print(data_LC.isnull().sum())
```

<h3>Chuẩn hóa cột văn bản</h3>

```python
def clean_text_columns(df):
    text_columns = [
        'hinh_anh','thanh_phan', 'cong_dung', 'chong_chi_dinh', 
        'lieu_dung', 'luu_y', 'tac_dung_phu', 'bao_quan', 'mo_ta'
    ]
    
    def clean_text(text):
        if pd.isna(text) or text == 'nan':
            return np.nan
        
        text = str(text)
        text = BeautifulSoup(text, 'html.parser').get_text()   # Loại bỏ HTML tags
        text = re.sub(r'[\n\r\t]+', ' ', text)  # Loại bỏ ký tự xuống dòng và tab
        text = re.sub(r'[^\w\s\.,\-\(\)\[\]\/\+\%]', '', text, flags=re.UNICODE) # Loại bỏ ký tự đặc biệt không cần thiết
        text = re.sub(r'\s+', ' ', text)  # Loại bỏ khoảng trắng thừa
        text = text.strip()
        return text if text and text != '' else np.nan #  Trả về NaN nếu chuỗi rỗng sau khi làm sạch
    
    for col in text_columns:
        if col in df.columns:
            df[col] = df[col].apply(clean_text)

    return df

data_LC = clean_text_columns(data_LC)
```

```python
def normalize_price_column(df):   
    def clean_price(price):
        if pd.isna(price) or price == 'nan':
            return np.nan
        
        price = str(price)
        price = re.sub(r'[^\d,.]', '', price) # Loại bỏ các ký tự không phải số
        
        if ',' in price and '.' not in price:  # Thay thế dấu phẩy bằng dấu chấm (nếu là phân cách thập phân)
            price = price.replace(',', '.')
        elif ',' in price and '.' in price:
            
            price = price.replace(',', '') # Nếu có cả dấu phẩy và chấm, loại bỏ dấu phẩy (coi như phân cách hàng nghìn)
        
        try:
            return float(price)
        except:
            return np.nan
    
    if 'gia' in df.columns:
        print("Đang xử lý cột giá: gia")
        df['gia'] = df['gia'].apply(clean_price)
        print("Hoàn thành chuẩn hóa cột giá!")
    else:
        print("Không tìm thấy cột 'gia' trong dataset!")
    
    return df

data_LC = normalize_price_column(data_LC)

if 'gia' in data_LC.columns:
    print("\nThống kê cột giá:")
    print(data_LC['gia'].describe())
```

```python
def remove_duplicate_info(df):
    
    description_columns = [
        'hinh_anh', 'thanh_phan', 'cong_dung', 'chong_chi_dinh', 
        'lieu_dung', 'luu_y', 'tac_dung_phu', 'bao_quan', 'mo_ta'
    ]
    existing_desc_cols = [col for col in description_columns if col in df.columns]  
    print(f"Kiểm tra trùng lặp trong các cột: {existing_desc_cols}")
    
    def combine_unique_info(row): # Tạo cột tổng hợp thông tin không trùng lặp
        combined_info = {}
        
        for col in existing_desc_cols:
            if pd.notna(row[col]) and str(row[col]).strip():
                content = str(row[col]).strip()
                is_duplicate = False # Kiểm tra nội dung đã tồn tại chưa
                for existing_col, existing_content in combined_info.items():
                    if content.lower() in existing_content.lower() or existing_content.lower() in content.lower():
                        
                        if len(content) > len(existing_content): # Nếu trùng lặp, giữ lại nội dung dài hơn
                            combined_info[existing_col] = content
                        is_duplicate = True
                        break
                
                if not is_duplicate:
                    combined_info[col] = content
        return combined_info
   
    duplicate_count = 0 
    for idx, row in df.iterrows():
        unique_info = combine_unique_info(row)
        # Đếm số lượng trùng lặp được loại bỏ
        original_count = sum(1 for col in existing_desc_cols if pd.notna(row[col]) and str(row[col]).strip())
        new_count = len(unique_info)
        duplicate_count += (original_count - new_count)
        for col in existing_desc_cols:
            if col in unique_info:
                df.at[idx, col] = unique_info[col]
            else:
                df.at[idx, col] = np.nan
    
    print(f"Đã loại bỏ {duplicate_count} trường thông tin trùng lặp!")
    return df

data_LC = remove_duplicate_info(data_LC)
```

```python

def summarize_preprocessing(df):
    print(f"\nKích thước dữ liệu sau xử lý: {df.shape}")
    print(f"\nSố lượng giá trị thiếu theo cột:")
    missing_summary = df.isnull().sum()
    print(missing_summary[missing_summary > 0])
    
    print(f"\nCác cột số:")
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    print(numeric_cols)
    
    print(f"\nCác cột văn bản:")
    text_cols = df.select_dtypes(include=['object']).columns.tolist()
    print(text_cols)
    
    # Kiểm tra chất lượng dữ liệu
    print(f"\nChất lượng dữ liệu:")
    total_cells = df.shape[0] * df.shape[1]
    missing_cells = df.isnull().sum().sum()
    completion_rate = ((total_cells - missing_cells) / total_cells) * 100
    print(f"Tỷ lệ hoàn thiện dữ liệu: {completion_rate:.2f}%")
    
    return df

data_LC_processed = summarize_preprocessing(data_LC)

output_file = 'Longchau_data_processed.csv'
data_LC_processed.to_csv(output_file, index=False, encoding='utf-8')
print(f"\nĐã lưu dữ liệu đã xử lý vào file: {output_file}")

data_LC_processed 
```
