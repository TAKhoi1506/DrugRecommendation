```python
from bs4 import BeautifulSoup
import requests
import time
import csv
import os
```

```python
def get_all_danhmuc_links():
    base_url = "https://www.dieutri.vn"
    start_url = f"{base_url}/thuocgoc"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)' 
    }

    response = requests.get(start_url, headers=headers)
    response.raise_for_status()  

    soup = BeautifulSoup(response.text, 'html.parser')
    
    danhmuc_links = []
    for link in soup.select('div.list-content a.tit'):
        href = link['href'].strip()
        if href.startswith('http'):
            danhmuc_links.append(href)

    return danhmuc_links
```

```python
output_dir = 'danh_muc_links'
os.makedirs(output_dir,exist_ok=True)
```

```python
links = get_all_danhmuc_links()

txt_path = os.path.join(output_dir,'danh_muc_links.txt')
with open(txt_path,'w',encoding='utf-8') as f:
    for link in links:
        f.write(link + '\n')

print(f'Đã lưu thành công {len(links)} vào file: {txt_path}')
```

```python
def read_danhmuc_links(file_path):
    with open(file_path,'r',encoding = 'utf-8') as f:
        return[line.strip() for line in f if line.strip()]
       
```

```python
def get_links_from_danhmuc(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
    }

    response = requests.get(url,headers = headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')

    links_thuoc = []
    for a in soup.select('h2.tit-h a'):
        ten_thuoc = a.text.strip()
        link = a['href'].strip()
        links_thuoc.append((ten_thuoc,link))
    return links_thuoc
```

```python
def collect_all_thuoc_links(file_path):
    all_thuoc_links = []
    danhmuc_links = read_danhmuc_links(file_path)

    for url in danhmuc_links: 
        print(f'Đang tìm trong danh mục: {url}')
        try:
            DS_thuoc = get_links_from_danhmuc(url)
            all_thuoc_links.extend(DS_thuoc)
        except Exception as e:
            print(f'Lỗi khi lấy từ {url}: {e}')
        
    print(f'Tổng số thuốc đã lấy: {len(all_thuoc_links)}')
    return all_thuoc_links
```

```python
thuoc_links = collect_all_thuoc_links('danh_muc_links\danh_muc_links.txt')
output_dir = 'danh_muc_links'

for name,link in thuoc_links:
    print(f'{name} -> {link}')

#Lưu vào file txt
txt_path = os.path.join(output_dir,'DS_thuoc_links.txt')
with open(txt_path,'w',encoding='utf-8') as f:
    for name, link in thuoc_links:
        f.write(link + '\n')

print(f'Đã lưu thành công {len(thuoc_links)} vào file: {txt_path}')
```
