```python
import pandas as pd
import numpy as np
import re
from collections import Counter
```

```python
data_DT = pd.read_csv('D:\OU\DoAn\DrugRecommandation\crawlData_DieuTri\Thuoc_DieuTri_data.csv', encoding='utf-8')

print(f"Kích thước dữ liệu ban đầu: {data_DT.shape}")
print(f"Số cột: {len(data_DT.columns)}")
print(f"Danh sách cột: {data_DT.columns.tolist()}")

missing_values = data_DT.isnull().sum()
print(f"Số lượng giá trị thiếu theo cột:\n{missing_values}")
data_DT
```

```python
import unicodedata
def normalize_text(s):
    return unicodedata.normalize("NFC", s).lower().strip()
    
cols_chi_dinh = [ col for col in data_DT.columns if "chỉ định" in normalize_text(col) and "chống chỉ định" not in normalize_text(col) ]

cols_chong_chi_dinh = [ col for col in data_DT.columns if "chống chỉ định" in normalize_text(col) ]

cols_chi_dinh = list(set(cols_chi_dinh) - set(cols_chong_chi_dinh))

cols_lieu_dung = [col for col in data_DT.columns if 'liều dùng' in col.lower()]

cols_tac_dung_phu = [col for col in data_DT.columns if 'tác dụng phụ' in col.lower()]

cols_thanh_phan = [col for col in data_DT.columns if 'thành phần' in col.lower()]

cols_cach_dung = [col for col in data_DT.columns if 'cách dùng' in col.lower()]

cols_tuong_tac_thuoc = [col for col in data_DT.columns if 'tương tác thuốc' in col.lower()]

cols_than_trong = [col for col in data_DT.columns if 'thận trọng' in col.lower()]

cols_canh_bao = [col for col in data_DT.columns if 'cảnh báo' in col.lower()]

cols_mang_thai_cho_con_bu = [col for col in data_DT.columns if 'mang thai' in col.lower() or 'cho con bú' in col.lower()]

# Hiển thị số cột tìm được
print("Số cột tìm được theo từ khóa:")
print(f"Chỉ định: {len(cols_chi_dinh)} cột")
print(f"Chống chỉ định: {len(cols_chong_chi_dinh)} cột")
print(f"Liều dùng: {len(cols_lieu_dung)} cột")
print(f"Tác dụng phụ: {len(cols_tac_dung_phu)} cột")
print(f"Thành phần: {len(cols_thanh_phan)} cột")
print(f"Cách dùng: {len(cols_cach_dung)} cột")
print(f"Tương tác thuốc: {len(cols_tuong_tac_thuoc)} cột")
print(f"Thận trọng: {len(cols_than_trong)} cột")
print(f"Cảnh báo: {len(cols_canh_bao)} cột")
print(f"Mang thai và cho con bú: {len(cols_mang_thai_cho_con_bu)} cột")
```

```python
groups = {
    "1. Chỉ định": cols_chi_dinh,
    "2. Chống chỉ định": cols_chong_chi_dinh,
    "3. Liều dùng": cols_lieu_dung,
    "4. Tác dụng phụ": cols_tac_dung_phu,
    "5. Thành phần": cols_thanh_phan,
    "6. Cách dùng": cols_cach_dung,
    "7. Tương tác thuốc": cols_tuong_tac_thuoc,
    "8. Thận trọng": cols_than_trong,
    "9. Cảnh báo": cols_canh_bao,
    "10. Mang thai và cho con bú": cols_mang_thai_cho_con_bu,
}

print("=== CHI TIẾT CÁC CỘT ĐƯỢC NHÓM THEO TỪ KHÓA ===")
for group_name, cols in groups.items():
    print(f"\n{group_name}: {len(cols)} cột")
    if cols:
        for i, col in enumerate(cols, start=1):
            print(f"   {i}. {col}")
    else:
        print("Không tìm thấy cột nào")

total_grouped_cols = sum(len(cols) for cols in groups.values())
print("\n=== TỔNG KẾT ===")
print(f"Tổng số cột được nhóm: {total_grouped_cols}/{len(data_DT.columns)} cột")
print(f"Tỷ lệ cột được nhóm: {(total_grouped_cols/len(data_DT.columns))*100:.1f}%")

all_grouped_cols = sum(groups.values(), [])  # nối tất cả list lại
ungrouped_cols = [col for col in data_DT.columns if col not in all_grouped_cols]

print(f"\nCác cột không được nhóm ({len(ungrouped_cols)} cột):")
for i, col in enumerate(ungrouped_cols[:10], start=1):  # Chỉ hiển thị 10 cột đầu
    print(f"   {i}. {col}")
if len(ungrouped_cols) > 10:
    print(f"...và {len(ungrouped_cols)-10} cột khác")
```

```python
data_DT['Chỉ định'] = data_DT[cols_chi_dinh].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Chống chỉ định'] = data_DT[cols_chong_chi_dinh].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Tác dụng phụ'] = data_DT[cols_tac_dung_phu].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Tương tác thuốc'] = data_DT[cols_tuong_tac_thuoc].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Liều dùng'] = data_DT[cols_lieu_dung].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Cách dùng'] = data_DT[cols_cach_dung].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Thận trọng'] = data_DT[cols_than_trong].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Thành phần'] = data_DT[cols_thanh_phan].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Cảnh báo'] = data_DT[cols_canh_bao].apply(lambda row:'\n'.join(row.dropna()), axis=1)
data_DT['Mang thai và cho con bú'] = data_DT[cols_mang_thai_cho_con_bu].apply(lambda row:'\n'.join(row.dropna()), axis=1)


data_DT.dropna(axis=1, how='all', inplace=True) #Loại bỏ các cột có giá trị trống hoàn toàn

print(f"Kích thước sau khi gộp cột: {data_DT.shape}")
grouped_columns = ['Chỉ định', 'Chống chỉ định', 'Tác dụng phụ', 'Tương tác thuốc', 
                  'Liều dùng', 'Cách dùng', 'Thận trọng', 'Thành phần', 'Cảnh báo', 
                  'Mang thai và cho con bú']

data_DT
```

```python
data_DT.rename(columns={'ten_thuoc': 'Tên thuốc'}, inplace=True)
main_cols = ['Tên thuốc','link','Chỉ định', 'Chống chỉ định', 'Tác dụng phụ', 'Tương tác thuốc', 'Liều dùng', 'Cách dùng', 'Thận trọng', 'Thành phần', 'Cảnh báo', 'Mang thai và cho con bú']

existing_main_cols = [col for col in main_cols if col in data_DT.columns]
print(f"Các cột được giữ lại: {existing_main_cols}")

data_DT_cleaned = data_DT[existing_main_cols].copy()
print(f"Kích thước sau khi chọn cột chính: {data_DT_cleaned.shape}")

data_DT_cleaned
```

```python
def clean_text_data(df):
    def clean_text(text):
        if pd.isna(text) or text == 'nan':
            return np.nan
        text = str(text)
        text = re.sub(r'<[^>]+>', '', text) # Loại bỏ HTML tags
        text = re.sub(r'[\n\r\t]+', ' ', text)  # Chuẩn hóa khoảng trắng
        text = re.sub(r'\s+', ' ', text)
        # Loại bỏ ký tự đặc biệt (giữ dấu câu cơ bản)
        text = re.sub(r'[^\w\s\.,;:!?\-\(\)\[\]\/\+\%°]', '', text, flags=re.UNICODE)
        text = re.sub(r'\.{2,}', '.', text) # Chuẩn hóa dấu câu
        text = re.sub(r',{2,}', ',', text)
        text = text.strip()
        
        return text if text and len(text) > 3 else np.nan
    
    text_columns = [col for col in df.columns if col not in ['Tên thuốc', 'link']]
    
    print(f"Làm sạch {len(text_columns)} cột văn bản...")
    
    for col in text_columns:
        if col in df.columns:
            print(f"  - Đang làm sạch cột: {col}")
            df[col] = df[col].apply(clean_text)
    
    print("Đã hoàn thành việc làm sạch dữ liệu văn bản!")
    return df

data_DT_text_cleaned = clean_text_data(data_DT_cleaned)
```

```python
def finalize_dataset(df):
    initial_rows = len(df)# Xóa hàng không có tên thuốc
    df = df.dropna(subset=['Tên thuốc'])
    removed_rows = initial_rows - len(df)
    if removed_rows > 0:
        print(f"Đã xóa {removed_rows} hàng không có tên thuốc")
    

    empty_columns = df.columns[df.isnull().all()].tolist()# Xóa cột hoàn toàn trống
    if empty_columns:
        df = df.drop(columns=empty_columns)
        print(f"Đã xóa {len(empty_columns)} cột hoàn toàn trống: {empty_columns}")
    
    priority_columns = [
        'Tên thuốc', 'link', 'Chỉ định', 'Thành phần', 
        'Liều dùng', 'Cách dùng', 'Tác dụng phụ',
        'Chống chỉ định', 'Thận trọng', 'Cảnh báo',
        'Tương tác thuốc', 'Mang thai và cho con bú'
    ]
    
    existing_priority = [col for col in priority_columns if col in df.columns]
    other_columns = [col for col in df.columns if col not in existing_priority]
    final_order = existing_priority + other_columns
    
    df = df[final_order]
    
    print(f"Kích thước cuối cùng: {df.shape}")
    print(f"\nCác cột trong dataset cuối:")
    for i, col in enumerate(df.columns):
        data_count = df[col].count()
        missing_pct = df[col].isnull().sum() / len(df) * 100
        print(f"  {i+1:2d}. {col:<30} ({data_count:4d} giá trị, {missing_pct:.1f}% thiếu)")
    
    return df

data_DT_final = finalize_dataset(data_DT_text_cleaned)
```

```python
def summarize_preprocessing(df_original, df_final):
    print(f"\nSo sánh trước và sau:")
    print(f"Kích thước ban đầu: {df_original.shape}")
    print(f"Kích thước cuối cùng: {df_final.shape}")
    print(f"Giảm {df_original.shape[1] - df_final.shape[1]} cột")
    print(f"Giảm {df_original.shape[0] - df_final.shape[0]} hàng")
    
    print(f"\nChất lượng dữ liệu:")
    total_cells_final = df_final.shape[0] * df_final.shape[1]
    missing_cells_final = df_final.isnull().sum().sum()
    completion_rate = ((total_cells_final - missing_cells_final) / total_cells_final) * 100
    print(f"Tỷ lệ hoàn thiện dữ liệu: {completion_rate:.2f}%")
    
    print(f"\nCác cột chính trong dữ liệu cuối:")
    for i, col in enumerate(df_final.columns):
        data_count = df_final[col].count()
        print(f"{i+1:2d}. {col:<30} ({data_count:4d} giá trị)")
    
    return df_final

Data_DT_processed = summarize_preprocessing(data_DT, data_DT_final)
print(f"\nDữ liệu sau khi xử lý:")
Data_DT_processed
```
